```python
import pandas as pd
```

```python
chunk_size = 1000
# for chunk in pd.read_csv('E:\DE_Project\dataset\e_commerce_dataset\Amazon Sale Report.csv', chunksize=chunk_size):
#     #process_each chunk here

sales_df = pd.read_csv('E:\DE_Project\dataset\e_commerce_dataset\Amazon Sale Report.csv', nrows=100)
```

```python
sales_df.head(10)
```

```python
#get the columns of DataFrame
sales_df.columns
```

```python
# remove leading and trailing white-space from the column name
sales_df.columns = sales_df.columns.str.strip()
```

```python
# sales_df.columns.str.upper().str.replace(' ','_').str.replace('-', '_')
# or
sales_df.columns = sales_df.columns.str.upper().str.replace(r'[ -]', '_', regex=True)
```

```python
#dropping the columns from dataframe
sales_df.drop(columns=['PROMOTION_IDS', 'UNNAMED:_22'], inplace=True)
```

```python
sales_df.columns
```

```python
# type of value
type(sales_df.isnull())
type(sales_df.isnull().sum())
```

```python
#get the columns which has missing value in their rows
sales_df.columns[sales_df.isnull().sum() > 0]
```

```python
sales_df.fillna({'COURIER_STATUS':'Cancelled', 'CURRENCY': 'INR', 'FULFILLED_BY': 'Easy Ship'}, inplace=True)
```

```python
sales_df.head(100)
```

```python
sales_df.drop_duplicates()
```

```python
#get the colums data-types
sales_df.dtypes
```

```python
#removing white space from one columns data
sales_df['SALES_CHANNEL'] = sales_df['SALES_CHANNEL'].str.strip()
```

```python
for col_name in sales_df.columns:
    if sales_df[col_name].dtypes == 'object':
        sales_df[col_name] = sales_df[col_name].str.strip()
```

```python
#data-types-conversion
sales_df['SHIP_POSTAL_CODE'] = sales_df['SHIP_POSTAL_CODE'].astype(int)
```

```python
# Convert date value into standard date and it's type also
sales_df['DATE'] = pd.to_datetime(sales_df['DATE'],format='%m-%d-%y')
```

```python
sales_df.head(10)
```

```python
sales_df['STATUS'] = sales_df['STATUS'].replace('Shipped - Delivered to Buyer','Shipped')
```

```python
sales_df.to_csv('processed_amazon_sales_report.csv',index=False)
```
